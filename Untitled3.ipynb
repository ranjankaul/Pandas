```python
import pandas as pd
import numpy as np
df = pd.read_csv("~/Documents/pandas_timeseries.csv")
```

```python
df.head()
```

```python
df['Ts'] = pd.to_datetime(df['Ts'])
```

```python
df.info()
```

```python
Q_lookup = {1:'Q1',2:'Q2',3:'Q3',4:'Q4'}
M_lookup = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
df['year'] = df['Ts'].dt.year
df['month'] = df['Ts'].dt.month.apply(lambda x:M_lookup[x])
df['quarter'] = df['Ts'].dt.quarter.apply(lambda x:Q_lookup[x])
df['qy'] = df['quarter'].map(str)+"-"+df['year'].map(str)
```

```python
df.head()
```

```python
quarter_count= df['qy'].unique()
print(quarter_count,len(quarter_count))
```

```python
df_group = df.groupby('qy')
```

```python
for i in range(len(quarter_count)):
    a = df_group.get_group(quarter_count[i])
    print(a)
    #call the CP function!
    
    
```
